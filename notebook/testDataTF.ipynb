```python
import pandas as pd

path = r"C:\Users\FrancoPC\OneDrive\Desktop\TPO_CienciaDeDatos\data\processed\tm_values_GB1_2024_latest.csv"
df = pd.read_csv(path)

display(df.head(10))

print("Filas:", len(df))
n_clubs = df["club_name"].nunique()
print("Clubes únicos:", n_clubs, "→ ¿20?", n_clubs == 20)
print("Nulos player_name:", df["player_name"].isna().sum())
print("Nulos market_value_eur:", df["market_value_eur"].isna().sum())
print("Duplicados (player_name, club_name):", df.duplicated(subset=["player_name","club_name"]).sum())

print("\nDescribe market_value_eur:")
print(df["market_value_eur"].describe())

print("\nTop 10 valores de mercado:")
print(df.sort_values("market_value_eur", ascending=False).head(10)[["player_name","club_name","market_value_eur"]])

print("\nJugadores por club (top 10):")
print(df["club_name"].value_counts().head(10))
```
