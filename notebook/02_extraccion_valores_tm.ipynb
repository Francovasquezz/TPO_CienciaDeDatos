```python
import LanusStats as ls
import pandas as pd
import time



# ==============================================================================
# 1. CONFIGURACI√ìN: MODIFICAR ESTOS VALORES PARA CADA EQUIPO
# ==============================================================================

# Ejemplo URL: https://www.transfermarkt.com/club-atletico-river-plate/startseite/verein/209
TEAM_NAME = "CA River Plate"
TEAM_ID = "209"
SEASON = "2024"

# ==============================================================================
# 2. EXTRACCI√ìN (C√ìDIGO CORREGIDO)
# ==============================================================================


print(f"--- Extrayendo jugadores para: {TEAM_NAME} (ID: {TEAM_ID}) ---")

try:
    tm = ls.Transfermarkt()
    df_players = tm.scrape_players_for_teams(
        team_name=TEAM_NAME,
        team_id=TEAM_ID,
        season=SEASON
    )

    if df_players is not None and not df_players.empty:
        # !! CORRECCI√ìN: Usamos los nombres de columna en espa√±ol !!
        columnas_originales = ['Jugadores', 'Valor de mercado']
        
        # Verificamos que las columnas existan antes de seleccionarlas
        if all(col in df_players.columns for col in columnas_originales):
            df_final = df_players[columnas_originales].copy()
            
            # Renombramos a nombres consistentes en ingl√©s para nuestro CSV final
            df_final.rename(columns={'Jugadores': 'Player', 'Valor de mercado': 'MarketValueEUR'}, inplace=True)
            
            print(f"\n‚úÖ ¬°√âxito! Se extrajeron datos de {len(df_final)} jugadores.")
            
            # --- Mostramos la tabla resultante ---
            print("--- Tabla de Jugadores y Valor de Mercado ---")
            display(df_final)

            # --- Copiamos la tabla al portapapeles en formato CSV para Excel ---
            print("\nüìã Datos listos para copiar y pegar (formato CSV con punto y coma):")
            print(df_final.to_csv(index=False, sep=';'))

        else:
            print("\n‚ùå Error: Las columnas esperadas ('Jugadores', 'Valor de mercado') no se encontraron en el resultado.")
            print("   Columnas recibidas:", df_players.columns.tolist())

    else:
        print(f"\n‚ö†Ô∏è No se encontraron datos para {TEAM_NAME}.")

except Exception as e:
    print(f"\n‚ùå Ocurri√≥ un error inesperado extrayendo datos: {e}")
```
